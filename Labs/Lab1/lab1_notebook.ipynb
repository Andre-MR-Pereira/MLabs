```python
import numpy as np
import matplotlib.pyplot as plt
from lab1 import poly_fit, poly, poly_vector, SSE,indic
```

```python
x = np.load('./data1_x.npy')
y = np.load('./data1_y.npy')

a = poly_fit(x,y,2)

x_ = np.linspace(-1, 1, 101)
y_ = poly_vector(a, x_)

indic(SSE(y, poly_vector(a, x)),a)

plt.scatter(x,y)
plt.plot(x_,y_)
plt.show()
```
